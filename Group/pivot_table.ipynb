#### Pivot table is used to summarize data and use aggregate functions

Parameters Explained

1.values:
The column(s) to aggregate.
Only numeric data can be aggregated.

2.index:
Defines the row labels for the pivot table.
The data will be grouped based on this column.
columns:
Defines the column labels for the pivot table.

3.aggfunc:
The aggregation function used for summarization (default is 'mean').
Common functions: 'sum', 'mean', 'count', 'min', 'max', etc.
You can also provide custom aggregation functions.

4.fill_value:
Specifies a value to replace missing data (NaN).

5.margins:
Adds a totals row and column labeled 'All' (default is False).

6.dropna:
Removes columns where all values are NaN.

7.sort:
Whether to sort the index and columns (default is True).
