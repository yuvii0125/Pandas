# Filling the Null Values

```python
df2 = df.fillna(value = 0)
df2.head()
```

```python
df2.isnull().sum().sum()
```

```python
df3 = df.fillna(5)
df3.head()
```

```python
df3.isnull().sum().sum()
```

#### Filling the null values with the next value

```python
df4 = df.bfill()  # Backward fill
df4.head()
```

#### Filling the null values with the Previous value

```python
df5 = df.ffill()
df5.head()
```

#### with axis we can apply it to row or column

```python
df6 = df.bfill(axis=1)
df6.head()
```

Same for the ffill

#### Filling Different values in null in different columns 

```python
df8 = df.fillna({'society':'abcd','balcony' :'defg'})
df8.head()
```

## Filling null value with mean , median and mode

```python
df9 = df.fillna(value=df['balcony'].mean())
df9.tail()
```

```python
df10 = df.fillna(value=df['balcony'].median())
df10.tail()
```

```python
df11 = df.fillna(value=df['balcony'].mode()[0])
df11.tail()
```

```python
df12 = df.fillna(value=df['balcony'].min())
df12.tail()
```

```python
df13 = df.fillna(value=df['balcony'].max())
df13.head()
```

```python
# Fill NaN in column 'A' with the mean
# df['A'] = df['A'].fillna(df['A'].mean())

# Fill NaN in column 'B' with forward fill
## df['B'] = df['B'].fillna(method='ffill')
```

## Dropping NaN

#### It will remove all the rows which contain even 1 value

```python
df14 = df.dropna()
df14
```

#### how parameter in dropna It will have 2 values any and all

for all it will remove a row if all values are NaN

```python
df15 = df.dropna(how='all')
df15
```

```python
df16 = df.dropna(how='any')
df16
```

#### also dropna have axis parameter , in which default is 0 mean row is going to removed but for axis=1 the columns will be removed

```python
df17 = df.dropna(axis=1)
df17.head()
```

#### thres parameter

thresh = 2 , Drop rows where less than 2 non-NA values exist

```python
df18 = df.dropna(thresh=2)
df18
```
