# __Filtering Rows and Columns__

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r'../Datasets/covid_19.csv')
df.head()
```

## Row filter mask

all rows where Lat is greater than 40

```python
row_filter_mask = df['Lat'] > 40
```

```python
df.loc[row_filter_mask,:]
```

We can use multiple row conditions with & and | and == operator

```python
mask1 = df['Lat'] > 50
mask2 = df['Long'] < 0
```

```python
df.loc[(mask1 | mask2), :]
```

```python
df.loc[(mask1 & mask2), :]  
```

# __Column Filter mask__

```python
col_mask = df.columns.str.startswith('L')
df.loc[:,col_mask]
```

```python
col_mask #Always give boolean array
```

# __Filtering Both Rows and Columns__

```python
filtered_data = df.loc[df['Lat'] > 25, ['Lat', 'Long']]
print(filtered_data)
```
