# __Membership Filtering__

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../Datasets/Churn.csv')
df.head()
```

# isin() method

Syntax : DataFrame[column].isin(values)


.isin(values) checks if each element in the Series or DataFrame is present in the values.  <br />
values can be a list, set, or any iterable. <br />
It returns a boolean Series, which can be used for filtering. <br />

```python
filter = df.state.isin(['NY', 'NJ'])
df[filter].head()
```

~ filter

only shows values except NY and NJ

```python
filter = ~df.state.isin(['NY', 'NJ'])
df[filter].head()
```

# between()

Checks if values fall between a range (inclusive by default).



Syntax : Series.between(left, right, inclusive="both")

```python
filter = df['account length'].between(50,100)
df.loc[filter,:].head()
```

# any() and all()

any() : Checks if any value in a DataFrame or Series evaluates to True.

If any of one is True(1 or positive) returns True , else false

```python
df = pd.DataFrame({'A': [0, 0, 1], 'B': [0, 1, 0]})
print(df['A'].any())  
print(df.any(axis=1))  
```

all() : Checks if all values in a DataFrame or Series evaluate to True.

Series.all() <br/>
DataFrame.all(axis=0)

```python
print(df['A'].all())  
print(df.all(axis=1))  
```

It Have default => axis = 'rows' ==>  returns result for every column <br />
axis = 'column' ==> return results for every row

# __Query__

The .query() method in Pandas provides a concise way to filter rows based on logical conditions written as a string.

```python
df = pd.read_csv('../Datasets/Churn.csv')
df.head()
```

```python
filter = df.query('state in ["NY", "NJ"]')
filter.head()
```

```python
df.query('`account length` >= 50 and `account length` <= 100').head()
```

We can use python variable also in query

```python
min_day_calls = 100
set = ('KS', 'OH', 'NV')
filter = df.query('`total day calls` > @min_day_calls and state in @set')
filter.head()
```

# __Eval__ 

The .eval() method is used for evaluating mathematical expressions or logical operations on columns in a DataFrame. It provides a faster alternative to direct computations using Python syntax.

DataFrame.eval(expr, inplace=False, engine='numexpr')


```python
df.eval('minutes_per_call = `total day minutes` / `total day calls`', inplace=True)
df.head()
```
