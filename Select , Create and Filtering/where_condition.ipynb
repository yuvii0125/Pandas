# __Where Condition__

DataFrame.where(cond, other=np.nan, inplace=False, axis=None)


cond: A condition (boolean mask). Elements that satisfy the condition are kept; others are replaced. <br/>
other: The value to replace elements that donâ€™t meet the condition (default is NaN).<br/>
inplace: If True, modifies the original DataFrame/Series. Defaults to False.<br/>
axis: Determines whether the condition is applied row-wise (axis=0) or column-wise (axis=1).<br/>

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r'../Datasets/Student_Depression_Data.csv')
df.head()
```

```python
mask = df.where(df['City'] ==  'Ahmedabad').fillna(0)
print(mask)
```

```python
mask = df.where(df['City'] ==  'Ahmedabad').dropna()
print(mask)
```
