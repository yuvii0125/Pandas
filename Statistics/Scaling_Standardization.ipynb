### Scaling means changing the range of data 

# 1. Z-Score Scaling (Standardization) <br />
Z-score scaling, also known as Standardization, transforms the data such that it has a mean of 0 and a standard deviation of 1. This technique is useful when the data has varying scales and is sensitive to outliers. The formula is:

### z = (x - μ) / €


x is the data point  <br />
μ = mean of the feature  <br />
£ = the standerd deviation of the feature 

```python
import pandas as pd
import numpy as np

data = pd.DataFrame({'A': [1, 2, 3, 4, 5], 'B': [10, 20, 30, 40, 50]})

# Z-Score Scaling (Standardization)
data_standardized = (data - data.mean()) / data.std()
print(data_standardized)
```

```python
df = pd.read_csv(r'../Datasets/hungary_chickenpox.csv')
df.head()
```

```python
df.describe()
```

```python
def z_score(x):
    return (x - np.mean(x))/(np.std(x))
```

```python
df1 = df.drop('Date',axis='columns').apply(z_score,axis= 'rows')
df1
```

```python
df1.describe()
```

## 2. Min-Max Scaling
Min-Max scaling transforms the data to fit within a specific range, typically between 0 and 1

x' = x - min(x) / max(x) - min(x)

```python
def min_max(x):
    return (x - np.min(x))/(np.max(x) - np.min(x))
```

```python
df2 = df.drop('Date',axis='columns').apply(min_max,axis= 'rows')
df2
```

```python
df2.describe()
```

## 3. Maximum Absolute Scaling
Maximum Absolute Scaling scales the data by dividing each feature by its maximum absolute value. This results in a scaled range of 
[−1,1]for each feature,

x' = x / max(x)

```python
def max_scalling(x):
    return x/np.abs(np.max(x))
```

```python
zdf1 = df.drop('Date',axis='columns').apply(max_scalling,axis= 'rows')
zdf1
```

```python
zdf1.describe()
```

## 4. Robust Scaling <br />
Robust scaling is more robust to outliers. It scales data by removing the median and scaling according to the interquartile range (IQR)

x' = x - median(x) / Q3 - Q1 <br />
where Q1 = 25th percentile and Q3 = 75th percentile

```python
def robust_scalling(x):
    return (x - np.median(x))/(x.quantile(q=0.75) - x.quantile(0.25) )
```

```python
df4 = df.drop('Date',axis='columns').apply(robust_scalling,axis= 'rows')
df4
```

```python
df4.describe()
```
