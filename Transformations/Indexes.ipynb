```python
import pandas as pd

# Create a DataFrame
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})

# Insert a new column 'C' at position 1 (between 'A' and 'B')
df.insert(1, 'C', [7, 8, 9])
print(df)
```

```python
## General function to move a column 
```

```python
def move_column(df, column_name, new_position):
    # Pop the column and insert it at the new position
    column = df.pop(column_name)
    df.insert(new_position, column_name, column)
    return df
```

```python
df = pd.read_csv(r'../Datasets/Student_Depression_data.csv')
df.head()
```
