One-hot encoding is a method for converting categorical data into a binary matrix (a series of 0s and 1s), making it easier to use in machine learning models. 

```python
import pandas as pd
```

```python
data = {
    "City": ["London", "Paris", "New York", "London", "Paris"],
    "Temperature": [21, 24, 19, 22, 23]
}

df = pd.DataFrame(data)

# Perform one-hot encoding on the "City" column
encoded_df = pd.get_dummies(df, columns=["City"])

print(encoded_df)
```
