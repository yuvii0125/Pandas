# Removing Duplicates

removing duplicates items based on one or more columns is another often used dara cleaning step

1. Keep first occurence only <br />
2. Keep Last occurencee only <br />
3. Remove all if there's more than one occurence 

Syntax : DataFrame.drop_duplicates(subset=None, keep='first', inplace=False, ignore_index=False)


Parameters: <br />
subset: Columns to consider for identifying duplicates. Default is None (all columns).  <br />
keep: Determines which duplicate to keep: <br />
'first' (default): Keeps the first occurrence. <br />
'last': Keeps the last occurrence. <br />
False: Removes all duplicates. <br />
inplace: If True, modifies the DataFrame in place. Default is False. <br />
ignore_index: If True, resets the index of the resulting DataFrame. Default is False. <br />

```python
import pandas as pd
```

```python
df = pd.read_csv(r'../Datasets/Churn.csv')
df.head()
```

```python
df.drop_duplicates('state',keep='first',inplace=False).head()
```

```python
df.drop_duplicates('state',keep='last',inplace=False).head()
```

```python
df.drop_duplicates('state',keep=False,inplace=False).head()
```
