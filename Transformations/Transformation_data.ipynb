
## Assign , apply and map functions 


## 1.assign  <br />
Used to add or modify columns in a DataFrame.  <br />
Creates a new DataFrame without modifying the original one. <br />
You can pass a function or a value to create new columns.

### Syntax : DataFrame.assign(new_col_name=function_or_value)


```python
import pandas as pd

df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})

# Add a new column 'C' as the sum of 'A' and 'B'
new_df = df.assign(C=df['A'] + df['B'])

print(new_df)
```

## 2.map  <br />
Used for element-wise transformations on an entire DataFrame. <br />
Applies a function to each individual element of the DataFrame.  <br />

```python
import pandas as pd

df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})

# Apply a function to each element in the DataFrame
result = df.map(lambda x: x ** 2)

print(result)
```

## 3. apply  <br />

Can be used for row-level, column-level, or element-wise transformations, depending on the axis specified.<br />

DataFrame.apply(function, axis=1)  # Row-wise (axis=1)  <br />
Series.apply(function)             # For Series  <br />
