### __Displaying all uniques__

.unique() :- Gives unique values in particular column

```python
df['Degree'].unique()
```

.nunique() :- Gives No. of  unique values in particular column

```python
df['Degree'].nunique()
```

.value_counts() :- number of occurences of each item

```python
df['Degree'].value_counts() #if we give normalize=True, it will give the percentage of each category
```

```python
df['Age'].value_counts(normalize=True)
```

```python
df.nlargest(7,'Age') 
```
